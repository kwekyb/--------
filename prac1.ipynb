```python
import statistics

FN = "Darius-13-100m-Fly.txt"
FOLDER = "swimdata/"

swimmer, age, distance, stroke = fn.removesuffix(".txt").split("-")

times = []
with open (FOLDER + FN) as file:
    lines = file.readlines()
    for line in lines:
        # print(line)
        times.append(line.strip().split(","))
# print(times)

converts = []
aves = []
maxs = []
mins = []
max_time = 0
min_time = 0

for ts in times:
    for t in ts:
        minutes, rest = t.split(":")
        seconds, hundredths = rest.split(".")
        converts.append((int(minutes) * 60 * 100) + (int(seconds) * 100) + int(hundredths))
    
    average = statistics.mean(converts)
    max_time = max(converts)
    maxs.append(max_time)
    min_time = min(converts)
    mins.append(min_time)
    
    mins_secs, hundredths = str(round(average / 100, 2)).split(".")
    mins_secs = int(mins_secs)
    minutes = mins_secs // 60
    seconds = mins_secs - minutes * 60
    average = str(minutes) + ":" + str(seconds) + "." + hundredths
    aves.append(average)

maxf = []    
for ma in maxs:
    mins_secs, hundredths = str(round(ma / 100, 2)).split(".")
    mins_secs = int(mins_secs)
    minutes = mins_secs // 60
    seconds = mins_secs - minutes * 60
    max_time_str = str(minutes) + ":" + str(seconds) + "." + hundredths
    maxf.append(max_time_str)

minf = []
for mi in mins:
    mins_secs, hundredths = str(round(mi / 100, 2)).split(".")
    mins_secs = int(mins_secs)
    minutes = mins_secs // 60
    seconds = mins_secs - minutes * 60
    min_time_str = str(minutes) + ":" + str(seconds) + "." + hundredths
    minf.append(min_time_str)

print(maxf)
print(minf)

i = 0
for avg in aves:
    print(times[i])
    print(f"[{i}번] {avg}")
    print(f"최고 기록 : {maxf[i]}, 최저 기록: {minf[i]}")
    i = i + 1
```

```python
import statistics

# 설정 및 데이터 분리
FN = "Darius-13-100m-Fly.txt"
FOLDER = "swimdata/"
swimmer, age, distance, stroke = FN.removesuffix(".txt").split("-")

# 밀리초(백분의 1초)를 시계 형식(M:S.cc)으로 변환하는 함수
def format_time(hundredths_total):
    total_seconds = hundredths_total / 100
    minutes = int(total_seconds // 60)
    seconds = int(total_seconds % 60)
    hundredths = int(round((total_seconds - int(total_seconds)) * 100))
    return f"{minutes}:{seconds:02d}.{hundredths:02d}"

# 파일 읽기 및 처리
with open(FOLDER + FN) as file:
    lines = file.readlines()

for i, line in enumerate(lines):
    raw_times = line.strip().split(",")
    
    # 각 줄의 시간을 백분의 1초 단위 정수로 변환
    converts = []
    for t in raw_times:
        minutes, rest = t.split(":")
        seconds, hundredths = rest.split(".")
        converts.append((int(minutes) * 6000) + (int(seconds) * 100) + int(hundredths))
    
    # 통계 계산
    avg_val = statistics.mean(converts)
    max_val = max(converts)
    min_val = min(converts)
    
    # 결과 출력
    print(f"\n{raw_times}")
    print(f"[{i}번] 평균: {format_time(avg_val)}")
    print(f"최고 기록: {format_time(min_val)}, 최저 기록: {format_time(max_val)}")
```

파이썬의 2중 리스트(Nested List)는 행과 열을 가진 표(Table) 형태의 데이터를 다루는 가장 기본적인 구조입니다. 이를 능숙하게 다루기 위해 반복 코딩과 암기가 필요한 핵심 패턴 3가지를 정리해 드릴게요.
1. 시각적 구조 이해
2중 리스트를 다룰 때는 항상 '리스트 안의 리스트'라는 점을 기억해야 합니다. matrix[i][j]에서 i는 큰 행(Row)을, j는 그 안의 개별 요소(Column)를 가리킵니다.

2. 반복 코딩으로 익혀야 할 핵심 패턴
① 모든 요소 순회하기 (기본형)
가장 많이 쓰이는 패턴입니다. 외워두면 중첩 루프의 감이 잡힙니다.
Python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
 
# 인덱스를 사용하는 방식 (수정이 필요할 때 유용)
for i in range(len(matrix)):          # 행 반복
    for j in range(len(matrix[i])):   # 열 반복
        print(matrix[i][j], end=" ")
    print() # 행이 끝나면 줄바꿈

② 리스트 컴프리헨션 (심화형)
파이썬다운(Pythonic) 코드를 작성하기 위해 이 패턴은 통째로 익숙해지는 것이 좋습니다. 2차원 리스트를 초기화하거나 변환할 때 필수입니다.
3x4 크기의 0으로 채워진 리스트 만들기:
Python
# [결과] * 반복횟수 형태
arr = [[0] * 4 for _ in range(3)]
1차원 리스트를 2차원으로 만들기 (변형 연습):
Python
flat = [1, 2, 3, 4, 5, 6]
# 2개씩 묶어서 2중 리스트 생성
nested = [flat[i:i+2] for i in range(0, len(flat), 2)]

③ 전치 행렬 (Transpose) 만들기
행과 열을 바꾸는 작업입니다. 코딩 테스트나 데이터 처리에서 자주 등장하는 논리입니다.
Python
# 행과 열 뒤집기
transposed = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
 
# zip을 활용한 마법 같은 한 줄 (이 패턴을 외우면 매우 편합니다)
transposed = list(map(list, zip(*matrix)))

Tip: 코드를 짤 때 머릿속으로 엑셀 시트를 그리거나, 종이에 $3 \times 3$ 격자를 그려놓고 인덱스 (0,0), (0,1)...을 따라가는 연습을 병행하면 훨씬 빨리 이해됩니다.
현재 어떤 종류의 데이터를 2중 리스트로 처리하려고 하시나요? 구체적인 데이터 예시(예: 점수 표, 좌표값 등)를 알려주시면 그에 맞는 맞춤형 연습 코드를 짜 드릴 수 있습니다.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
 
# 인덱스를 사용하는 방식 (수정이 필요할 때 유용)
for i in range(len(matrix)):          # 행 반복
    for j in range(len(matrix[i])):   # 열 반복
        print(matrix[i][j], end=" ")
    print() # 행이 끝나면 줄바꿈
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
 
# 인덱스를 사용하는 방식 (수정이 필요할 때 유용)
new_matrix = []
for items in matrix:          # 행 반복
    row = []
    for item in items:   # 열 반복
        row.append(item)
    new_matrix.append(row)
print(new_matrix) # 행이 끝나면 줄바꿈
```
